```
# Since we're using databricks, a spark context is automatically created for us

print(sc)
```

```
# Creating an rdd from the range

a = sc.parallelize(range(1,100))
print(a.collect())
print(type(a))
```

```
# getting the number of partitions of the rdd (a) above.

a.getNumPartitions()
```

```
# Specifying the number of partitions while creating the rdd

b = sc.parallelize(range(1,100),10)
b.getNumPartitions()
```

```
print(a.collect())
print(b.collect())
```

```
# Repartitioning the rdd
# This shuffles the data before partitioning

c = a.repartition(10)
c.getNumPartitions()
```

```
# Reducing the number of partitions
# Shuffles the data only if necessary

d = c.coalesce(4)
d.getNumPartitions()
```

```
# creating an rdd from the textfile

book = sc.textFile('dbfs:/FileStore/shared_uploads/ndutiv@gmail.com/Robert_Greene_John_Clark_Jord__ebooksread_com_-1.txt')
```

```
# Returns the first n elements of an rdd

book.take(5)
```

```
type(book)
```

```
# Returns the number of elements in an rdd

book.count()
```

```
# Returns the first element of an rdd or None if empty

book.first()
```

```
# flattens the rdd(nested structures) and returns an rdd with only top level elements

a = book.flatMap(lambda a: a.split(" "))
print(a.collect())
```

```
a.take(2)
```

```
# returns a list of tuples of every element with 1

b = a.map(lambda x:(x,1))
print(b.collect())
```

```
# Returns a list of tuples, each of which contains a key and the sum of the values for that key

c = b.reduceByKey(lambda x,y : (x+y))
```

```
output = c.collect()
print(output)
```

```
print(type(output))
print(type(c))
```

```
# Sorting the output by key

d = c.sortByKey(ascending=False).collect()
print(d)
```

```
# sort an RDD by the value of the second element in each tuple

e = c.sortBy(lambda x:x[1],ascending=False).collect()
print(e)
```

```
for (word,count) in e:
    print("%s : %i"%(word,count))
```
